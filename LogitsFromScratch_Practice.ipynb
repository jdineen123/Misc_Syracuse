```python
from sklearn import datasets
from sklearn.cross_validation import train_test_split
iris = sklearn.datasets.load_iris()



x = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.50, random_state=42)

number_features = x.shape[1]
number_classes = np.max(iris.target) + 1
num_obs = x.shape[0]
b = np.ones(number_classes)

print('Shape of x is: {}'.format(x.shape))
print('Shape of y is: {}'.format(y.shape))

print('No Obs {}'.format(num_obs))
print('No classes {}'.format(number_classes))
print('No Feats {}'.format(number_features))

#Weights vector is Rows = Features * Cols = Classes
w_shape = (number_features, number_classes)
w = np.zeros(w_shape)

w
```

```python
def _one_hot(y, n_labels, dtype):

    mat = np.zeros((len(y), n_labels))
    for i, val in enumerate(y):
        mat[i, val] = 1
    return mat.astype(dtype)


y_enc = _one_hot(y, 3, np.float)
print('Shape of y is: {}'.format(y_enc.shape))
y_enc
```

```python
#weights vector needs to have a row for each feature, and a column for each class
#We have 2 observations, and 3 classes
def softmax(z):
    return (np.exp(z.T) / np.sum(np.exp(z), axis=1)).T


def cross_entropy(output, y_target):
    return -np.sum(np.log(output) * (y_target), axis=1)


net_input = x.dot(w) + b
y_probs = softmax(net_input)
cost = []
for i in range(1, 10001):
    diff = y_probs - y_enc
    grad = np.dot(x.T, diff)/y.size
    learning_rate = .1
    w -= learning_rate * grad
    loss = cross_entropy(y_probs, y_enc)
    cost.append(loss)
    net_input = x.dot(w) + b
    y_probs = softmax(net_input)

for i, j in enumerate(cost):
    print('epoch {}, cost {}'.format((i + 1), (np.round(np.mean(j), 10))))
```

### Build a simple Class

```python
w
from sklearn.metrics import accuracy_score, confusion_matrix

test = X_test.dot(w) + b
probs = np.exp(test) / np.sum(np.exp(test))
preds = []
for i in probs:
    preds.append(np.argmax(i)) 

preds = np.asarray(preds)

accuracy_score(y_test, preds)
c = confusion_matrix(y_test, preds, labels = [0,1])
c
```
